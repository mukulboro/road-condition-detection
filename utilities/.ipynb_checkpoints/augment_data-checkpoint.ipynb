```python
import cv2 as cv
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
dir_to_dataset_folder = "D:/ML_project/model/road_condition_data"
```

```python
images = {}
count = 0
dir_to_image_folder = dir_to_dataset_folder+"/full_dataset"
for filename in os.listdir(dir_to_image_folder):
    count += 1
    img = cv.imread(os.path.join(dir_to_image_folder, filename))
    if img is not None:
        print(f"{count}:{filename}")
        images[filename] = img
print(f"Added {count} images to list.::::{len(images)}")
```

```python
dir_to_annotation = dir_to_dataset_folder + "/annotations.csv"
df = pd.read_csv(dir_to_annotation)

df = df.drop(["Unnamed: 0", "rating"],axis=1)
```

```python
def sp_noise(image,prob):
    '''
    Add salt and pepper noise to image
    prob: Probability of the noise
    '''
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
```

```python
new_df = pd.DataFrame(columns=["img_name", "has_pothole", "has_crack", "is_pitched"])
new_image_dir = "D:/ML_project/utilities/new_dataset/"
for index, row in df.iterrows():
    img_name = row["image_name"]
    has_potholes = row["has_potholes"]
    is_pitched = row["is_pitched"]
    has_cracks = row["has_cracks"]
    img_data = images[img_name]
    noise_image = sp_noise(img_data, 0.02)
    blur_image = cv.blur(img_data, (4,4))
    high_contrast_image = cv.convertScaleAbs(img_data, alpha=1.5, beta=1)
    low_contrast_image = cv.convertScaleAbs(img_data, alpha=0.3, beta=1)
    
    new_img_name = f"Image{index}.jpg"
    new_df = new_df.append({
        "img_name": new_img_name,
        "has_pothole": has_potholes,
        "has_crack": has_cracks,
        "is_pitched": is_pitched
    }, ignore_index=True)
    cv.imwrite(new_image_dir+new_img_name, img_data)
    
    noisy_image_name = f"Image{index}_noise.jpg"
    new_df = new_df.append({
        "img_name": noisy_image_name,
        "has_pothole": has_potholes,
        "has_crack": has_cracks,
        "is_pitched": is_pitched
    }, ignore_index=True)
    cv.imwrite(new_image_dir+noisy_image_name, noise_image)
    
    blur_image_name = f"Image{index}_blur.jpg"
    new_df = new_df.append({
        "img_name": blur_image_name,
        "has_pothole": has_potholes,
        "has_crack": has_cracks,
        "is_pitched": is_pitched
    }, ignore_index=True)
    cv.imwrite(new_image_dir+blur_image_name, blur_image)
    
    hc_image_name = f"Image{index}_hc.jpg"
    new_df = new_df.append({
        "img_name": hc_image_name,
        "has_pothole": has_potholes,
        "has_crack": has_cracks,
        "is_pitched": is_pitched
    }, ignore_index=True)
    cv.imwrite(new_image_dir+hc_image_name, high_contrast_image)
    
    lc_image_name = f"Image{index}_lc.jpg"
    new_df = new_df.append({
        "img_name": lc_image_name,
        "has_pothole": has_potholes,
        "has_crack": has_cracks,
        "is_pitched": is_pitched
    }, ignore_index=True)
    cv.imwrite(new_image_dir+lc_image_name, low_contrast_image)
    break

new_df.head(21)
```
